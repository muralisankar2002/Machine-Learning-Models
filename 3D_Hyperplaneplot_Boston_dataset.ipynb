```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
from sklearn.datasets import load_boston
dataset = load_boston()
```

```python
dataset
```

```python
df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
df.head()
```

```python
df['MEDV'] = dataset.target
```

```python
df.head()
```

```python
corr = df.corr()
```

```python
corr
```

```python
# To get largest three correlated attributes with 'MEDV'.
print(df.corr().abs().nlargest(3,'MEDV').index)
print(df.corr().abs().nlargest(3,'MEDV').values[:,13])
```

```python
x = pd.DataFrame(np.c_[df['LSTAT'], df['RM']], columns = ['LSTAT','RM'])
Y = df['MEDV']
```

```python
x_surface = np.arange(0, 40, 1) #---for LSTAT---
y_surface = np.arange(0, 10, 1) #---for RM---
x_surface, y_surface = np.meshgrid(x_surface, y_surface)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, Y)
```

```python
figure = plt.figure(figsize=(18,15))
ax = figure.add_subplot(111, projection = '3d')
ax.scatter(x['LSTAT'],x['RM'],Y, marker = '.')
ax.set_xlabel("LSTAT")
ax.set_ylabel('RM')
ax.set_zlabel('MEDV')
z = lambda x,y: (model.intercept_ + model.coef_[0] * x + model.coef_[1] * y)
ax.plot_surface(x_surface, y_surface, z(x_surface,y_surface),
rstride=1,
cstride=1,
color='None',
alpha = 0.4)
plt.show()
```
